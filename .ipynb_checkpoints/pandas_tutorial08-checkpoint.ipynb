# Starting out tutorial on datetime-index

```python
import pandas as pd
df=pd.read_csv('aapl.csv' , parse_dates=['Date'] , index_col='Date')
df.head(5) 
```

```python
df.index
```

It is used to filter data according to datetime index

```python
# this was the old method
df['2017-06']
```

```python
# this is the new method
df.loc['2017-06']
```

```python
# get the data of moth 6th and 5th
df.loc['2017-06':'2017-05']
```

```python
# get the average closing price of 5th month
df.loc['2017-05'].Close.mean()
```

```python
df.Close.resample('M').mean()

```
