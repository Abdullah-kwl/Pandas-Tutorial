```python
import pandas as pd
df=pd.read_csv('weather_by_cities2.csv')
df
```

```python
df.pivot(index='city' , columns='date')
```

```python
df.pivot(index='date' , columns= 'city')
```

```python
#  , values='humidity'
# only gave humidity column
df.pivot(index='date' , columns= 'city', values='humidity')
```

# pivioit table = it is used to sumerise data inside the table

```python
df2 = pd.read_csv('weather2.csv')
df2
```

```python
df2.pivot_table(columns='city' ,index='date')
```

```python
# this will sum the value
df2.pivot_table(columns='city' ,index='date' ,aggfunc=sum)
```
