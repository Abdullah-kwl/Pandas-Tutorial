# Cross_tabs in pandas

```python
import pandas as pd
df=pd.read_excel('survey.xls')
df
```

```python
df2=pd.crosstab(df.Nationality,df.Handedness)
df2
```

```python
df.Sex
```

```python
# try own
df3=pd.crosstab([df.Sex,df.Handedness],df.Nationality)
df3
```

```python
df4=pd.crosstab(df.Sex,df.Handedness , margins=True)
df4
```

```python
import numpy as np
pd.crosstab(df.Sex, df.Handedness, values=df.Age, aggfunc=np.average)
```
